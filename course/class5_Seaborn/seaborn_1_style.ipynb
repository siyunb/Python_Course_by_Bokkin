# seaborn
基于matplotlib，用更简单的调用呈现更多美图~鱼与熊掌兼得~
>If matplotlib “tries to make easy things easy and hard things possible”, seaborn tries to make a well-defined set of hard things easy too.


Seaborn跟matplotlib最大的区别就是它的默认绘图风格和色彩搭配都具有现代美感，其实是在matplotlib的基础上进行了更高级的API封装，让你能用更少的代码去调用 matplotlib的方法，从而使得作图更加容易，在大多数情况下使用seaborn就能做出很具有吸引力的图，而使用matplotlib就能制作具有更多特色的图。应该把Seaborn视为matplotlib的补充，而不是替代物。

## 风格管理
先来看看matplotlib画出来的图啥样子~

```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
np.random.seed(sum(map(ord, "aesthetics"))) # 给每一个任务单独分配随机种子
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7): # up to but not include
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

再来看看seaborn画出来滴
* 色彩柔和
* 灰色背景
* 白色网格

```python
import seaborn as sns
sns.set_style("dark")
sinplot()
```

* 风格参数设置
* 尺度参数设置

### 风格参数设置
* darkgrid/whitegrid：更关注数值
* dark/white：更关注整体趋势。white 和 ticks两个风格都能够移除顶部和右侧的不必要的轴脊柱。通过matplotlib参数是做不到这一点的，但是你可以使用seaborn的despine()方法来移除它们
* ticks：强化刻度

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
sns.despine()
```

```python
sinplot()
sns.despine(offset=50)
```

### 尺度参数设置

```python
sns.set_context("paper")
sinplot()
```

```python
sns.set_context("notebook")
sinplot()
```

```python
sns.set_context("talk")
sinplot()
```

```python
sns.set_context("poster")
sinplot()
```

## 颜色管理
### 分类调色板
* 适用于分类描述的数据，而且分类之间无序
* hls颜色系统，色调+亮度+饱和度（hue + luminance + saturation）
* husl颜色系统，hls中不同颜色存在视差，eg.黄色偏亮，蓝色偏暗，husl使得更加平衡

```python
# 恢复默认设置
sns.set()
current_palette = sns.color_palette()
sns.palplot(current_palette)
# 某一种调色板
sns.set_palette(sns.color_palette("hls", 10))
current_palette = sns.color_palette()
sns.palplot(current_palette)
#另一种调色板
sns.set_palette(sns.color_palette("husl", 10)) # http://www.hsluv.org
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

### 序列调色板
* 主要通过亮度和饱和度区分有序分类
* 关注单级数据

```python
sns.palplot(sns.color_palette("Greens",10))
```

### 分叉调色板
* 关注两级数据

```python
sns.palplot(sns.color_palette("autumn",11))
```
