## 什么是Bokeh
* 基于Web浏览器呈现的交互式可视化
* 多种语言支持

## Bokeh的优势
* 简单的coding快速创建“复杂”图表
* 提供到各种媒体的输出，HTML、Notebook、服务器、flash、django
* 可视化成果的转化，matplotlib、seaborn、ggplot
* 交互式可视化实现

## Bokeh的挑战
* 快速迭代带来的不稳定性
* 与D3.js相比可视化选项相对较少

Bokeh用于快速开发原型产品，D3.js用于生产阶段产品开发

## Bokeh的组成
* charts：高级接口，简单快速地建立复杂的统计图表（transform to bkcharts package）
* plotting：中级接口，构建各种视觉符号
* models：低级接口，为应用程序开发人员提供最大的灵活性

### bokeh.models
* BokehJS：JavaScript库，负责所有绘图和交互的底层实现
* Json生成：应用相关库，用于生成json

### bokeh.plotting
* 专注于图形的绘制，类似matplotlib

```python
from bokeh.plotting import figure, output_notebook, show

# create a Figure object
p = figure(plot_width=300, plot_height=300, tools="pan,reset,save")

# add a Circle renderer to this figure
p.circle([1, 2.5, 3, 2], [2, 3, 1, 1.5], radius=0.3, alpha=0.5)

# specify how to output the plot(s)
output_notebook()

# display the figure
show(p)
```
