## 常用设置
* 获得当前图表对象
fig = matplotlib.pyplot.gcf() # gcf = get current figure
* 常用图表对象设置
fig.set_size_inches(10,10), matplotlib.pyplot.figure(figsize=(14,7))
* 保存当前图表对象
matplotlib.pyplot.savefig("figure1.jpg")

```python
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import os
import scipy
# 包含不同的pdf
from scipy import stats
# 让图片显示在notebook中
%matplotlib inline
# plt.show(), plt.cla(), plt.clf(), plt.close()
os.chdir("D:/3dsll")               #设定工作路径
```

```python
plt.figure(figsize=(10,5))
x = np.arange(-5,5,0.01) # 生成数字序列
y = stats.norm.pdf(x,0,1) # loc = mean, scale = sigma
plt.plot(x,y)
plt.savefig("figure1.png") # 不支持jpg
plt.figure(figsize=(5,5))
plt.savefig("figure2.png") # 思考为什么figure2是空的？
fig = plt.gcf()
fig.set_size_inches(5,5)
plt.savefig("figure3.png") # 思考为什么figure3是空的？
plt.plot(x,y)
fig = plt.gcf()
fig.set_size_inches(5,5)
plt.savefig("figure4.png")
```

## 常用图形
* 曲线图：matplotlib.pyplot.plot(data)
* 灰度图：matplotlib.pyplot.hist(data)
* 散点图：matplotlib.pyplot.scatter(data)
* 箱式图：matplotlib.pyplot.boxplot(data)

```python
x = np.arange(-5,5,0.1)
y = x ** 2
plt.plot(x,y)
```

```python
x = np.random.normal(size=1000)
plt.hist(x, bins=10)
```

```python
x = np.random.normal(size=1000)
y = np.random.normal(size=1000)
plt.scatter(x,y)
```

```python
plt.boxplot(x)
```

#### 箱式图科普
* 上边缘（Q3+1.5IQR）、下边缘（Q1-1.5IQR）、IQR=Q3-Q1
* 上四分位数（Q3）、下四分位数（Q1）
* 中位数
* 异常值
* 处理异常值时与$3\sigma$标准的异同：统计边界是否受异常值影响、容忍度的大小

btw：$\sigma \Rightarrow 68.2\%$，$2\sigma \Rightarrow 95.4\%$，$3\sigma \Rightarrow 99.7\%$

## 常用样式
### 颜色/线型/标记
* pyplot.plot(x, color='g')
* pyplot.plot(x, linestyle='--')
* pyplot.plot(x, marker='+')
* pyplot.plot(x, 'ro-')
### 标题
* pyplot.title("xxxx")
### 坐标轴标签
* pyplot.xlabel("xxxx")
* pyplot.ylabel("xxxx")
### 图注
* pyplot.legend(loc="best")配合pyplot.plot(label="xxxx")
### 网格
* pyplot.grid(color="xxx", linestyle="-", alpha=0.5)

```python
x = np.arange(-5,5.5,0.5)
y = x ** 2
plt.plot(x,y,color='g')
```

```python
plt.plot(x,y,linestyle='--')
```

```python
plt.plot(x,y,marker='+')
```

```python
plt.plot(x,y,'ro--')
```

```python
plt.title("style config")
plt.xlabel("x range")
plt.ylabel("y range")
plt.plot(x,y)
```

```python
y1 = x ** 2
y2 = x ** 3
plt.plot(x,y1,'r-',label='line 1')
plt.plot(x,y2,'b-',label='line 2')
plt.title("style config")
plt.xlabel("x range")
plt.ylabel("y range")
plt.legend(loc="best")
```

```python
plt.plot(x,y1,'r')
plt.plot(x,y2,'b')
plt.grid(color='k',linestyle='--',alpha=0.5)
```
